```python
import os
import time
import pandas as pd
import calendar
import matplotlib.pyplot as plt
#import matplotlib.pyplot as plt1


start_time = time.time()
name = "C:\\Users\\fa56892\\Desktop\\carac1"
L=[]
for f in os.listdir(name):    
    df = pd.read_csv(name+"\\"+f, sep = ";", encoding='latin-1')
    for i in df.index :
        if df.get_value(i, 'dep')==450 :
        #if df.get_value(i, 'an').isnull()!= "NaN" and df.get_value(i, 'mois')!=NaN  and df.get_value(i, 'jour')!=NaN: 
            L.append(calendar.weekday(int(df.get_value(i, 'an'))+2000, int(df.get_value(i, 'mois')), int(df.get_value(i, 'jour'))))

k=0
r=0
j=0
s=0
l=0
m=0
n=0
print(len(L))
for i in range(0,len(L)) : 
    if L[i]== 0:
        k+=1
    elif L[i]== 1:
        r+=1
    elif L[i]== 2:
        j+=1
    elif L[i]== 3:     
        s+=1
    elif L[i]== 4:     
        l+=1
    elif L[i]== 5:     
        m+=1 
    elif L[i]== 6:     
        n+=1
        
K1=[k,r,j,s,l,m,n]
print(K1)
labels = 'Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche'        
plt.pie(K1, labels=labels,
        autopct='%1.1f%%', startangle=140)
 
plt.axis('equal')
plt.title('RÃ©partition des accidents par jour de la semaine', fontweight = 'bold')
plt.show()
print("--- %s seconds ---" % (time.time() - start_time))

```
